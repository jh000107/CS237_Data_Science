### Hash functions (read carefully but no coding)
As you may recall from CS 112, hash functions map key values to buckets/bins in a hash table: the hash function appears to be spreading the keys uniformly randomly over the buckets, but in fact there is nothing random about it, since we can easily repeat the computation to find the key later. This is called <i>pseudo-random</i> behavior: the hash function is not random, but appears to be so unless you know the rule used to compute the hash function. 

The simplest hash functions use the Linear-Congruential Method, which you may remember from CS 112 (Google it if you are unsure whether you studied this); using prime numbers as multiplier and modulus is a good (but not perfect) way to simulate random behavior. The particular choices we will use here are from <a href="http://www.ams.org/journals/mcom/1999-68-225/S0025-5718-99-00996-5/S0025-5718-99-00996-5.pdf"> this </a> paper. 

For further details on the overall approach to generating pseudo-random numbers using the LC method,
see <a href="https://www.sciencedirect.com/topics/engineering/linear-congruential-generator">here</a>.

The basic idea is to start with a "seed value" and at each step, using the Linear Congruential Method
to generate the next value in the sequence, one at a time. So we will start with a seed value and successively apply the hash function to generate a series of pseudo-random numbers $n_1$, $n_2$, $n_3$, etc.

Supposing that our initial "seed" value is 1, we would have:

$n_0 = \text{1}$

$n_1 = \text{hash(1)}$

$n_2 = \text{hash(hash(1))}$

...

$n_k = \text{hash}^{k}\text{(1)}$
</blockquote>

NOTE that all of these numbers are in the range $[0 .. m_9)$, (since each one is the remainder after dividing by $m_9$).


### The following code prints a series of pseudo-random numbers in the range $[0 .. m_9)$. 

```python
a9 = 914334                     # remember to name global variables with problem number      
m9 = 2**22 - 3                  # You will submit your answer with these parameters

#a9 = 3                         # you will use these in part (b) just for testing
#m9 = 7

def hash(x):  
    return (a9 * x) % m9        # The Linear Congruenial Method

# Test it!
X9 = [231,45,123,87,133,123]     # for simple examples it is ok to use a Python list
for x in X9:
    print(hash(x))

print()
    
# Test it!
X9 = np.array([231,45,123,87,133,123])     # but use numpy arrays if you wish, they are mostly interchangable
for x in X9:
    print(hash(x))
    
```

###  Creating the sequence one random value at a time

Clearly the most efficient way to generate these numbers is to save the intermediate values,
and each step would involve hashing the old value, and storing the new value for the next step. 

The following code does that, to create the
basic random values that we will use for the rest of the lab. 

```python
next_value9 = 1     # just to create the variable which stores the number at each step

# seed next_value with the hash of (n+1)
def my_seed(n):
    global next_value9                  # if you are going to update a non-local variable you just declare it global
    next_value9 = hash(n+1)             # so values do not start with seed and are not 0       

# next_random_int() returns a random number generated by the hash function, in the range [0..m9) ; 

def next_random_int():
    global next_value9
    next_value9 = hash(next_value9)
    return next_value9                           
    
# Test it

my_seed(0)

print('\nRandom sequence of 100 random values from [0 ..',str(m9)+'):\n')
print( [ next_random_int()  for k in range(100) ] ) 
```

###  Part (B):  The Period Test

In addition to spreading the random numbers uniformly over a range, another important characteristic of a pseudo-random number generator is how many different values it can produce. If it can only produce 100 values
before it starts to repeat, then in an experiment with $10^5$ trials, it will repeat each value $10^3$ times,
which is hardly random behavior!   So we want sequences which are as long as possible!
    
The <i> period </i> of a generator is how long it takes before it starts to repeat the pseudo-random sequence, i.e., the number of unique values produced.  Why does it repeat?  Well, consider the case when $a9 = 3$ and $m9=7$ with an initial seed of 0:

$n_0=	3$<br>
$n_1=	2$<br>
$n_2=	6$<br>
$n_3=	4$<br>
$n_4=	5$<br>
$n_5=	1$<br>
$n_6=	3$<br>
$n_7=	2$<br>
$n_8=	6$

You can see that when we reach $n_6$ the sequence starts to repeat, at which point the sequence will be exactly the same again (why?). The period of this generator is $m-1 = 6$. No linear-congruential generator in the form we have given it can have a period larger than $m9-1$, because a 0 is problematic (why?), so this choice of a9 and m9 gives us a <i>full-period generator</i>.  Not all generators will have a full period, but we want as long a period as possible given $m9$, since we do not want pseudo-random numbers to repeat before the end of our random experiment-- this does not simulate random behavior very well!

###  Part (C):  The Spectral Test.

For this part, we would like you to do the <i>spectral test</i>. This is essentially the same as the
tests in HW 01 where we determined whether the probability function were equinumerous; however,
now we have floating-point numbers. 

In order to do this, since these are very close to behaving like real numbers: as with uncountable sample spaces, we really only can speak about intervals, not sample points. So we need to "bin" the numbers into some suitable intervals. For any number $k$ of equal-width bins over the range, if the probability is equinumerous, we would expect the probability of a number landing in a particular bin to be very close to $\frac{1}{k}$. 

<blockquote>
    There is a subtle problem with this test and its interaction with the period: if you generate exactly as many numbers as in the period, then by definition the bins will be almost exactly the same size, since the sequence generated all possible numbers in the range [0 .. m). (You can verify this by doing $m=4194301$ trials.) Hence we do not want to generate too much of the sequence to do the spectral test. We will therefore use only $10^5$ numbers, which represent

$$\frac{10^5}{4194301} \, =\, 0.0238$$
 
or about 2% of the period. 
</blockquote>

## Problem Eleven:  Display an Experimental Distribution

For this problem, provide code which will display the probability distribution for the experiment of running the "flip a coin until you get a head" experiment 100,000 times.

Hint:  when the sample space is countably infinite, you may generate some very large values!
For the purposes of this problem, define <code>upper_bound = 100</code>; when
generating values, stop when you get to the upper bound, thus you would return 100 if you got 99 tails in a row (since it does not matter what the next flip is). 

```python
# Solution
upper_bound = 100
num_trials = 100000
my_list = []
for i in range(num_trials):
    num_flips = 1
    while (num_flips < upper_bound):
        x = my_randint(2)
        if (x == 1):
            my_list.append(num_flips)
            break
        else:
            num_flips += 1

draw_distribution_from_outcomes(my_list)
```
