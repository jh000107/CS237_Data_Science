```python
# Here are some imports which will be used in code that we write for CS 237

# IMPORTANT:     DO NOT MAKE ANY OTHER IMPORTS WITHOUT DISCUSSING WITH PROFESSOR SNYDER

# Imports used for the code in CS 237

import numpy as np                # arrays and functions which operate on arrays, plus math functions
import matplotlib.pyplot as plt   # normal plotting 
import pandas as pd               # data management
import math                       # You may use the math library if you really want, but
                                  #    I recommend you use the numpy library for all mathematical operations.
                                  #    Examples of use are in the notebook numpy.ipynb on the class web site. 


from collections import Counter   # Essential for creating distributions from experiments
                                  # See hw00 for examples of how to use

# Useful code from HW 01 

# This draws a useful bar chart for the distribution of the list of integers in outcomes

def draw_distribution_from_outcomes(outcomes, title='Probability Distribution', my_xticks = [], f_size = (8,6)):
    plt.figure(figsize=f_size)
    num_trials = np.size(outcomes)
    X = range( int(min(outcomes)), int(max(outcomes))+1 )    # 
    freqs = Counter(outcomes)
    Y = [freqs[i]/num_trials for i in X]
    plt.bar(X,Y,width=1.0,edgecolor='black')
    if my_xticks != []:
        plt.xticks(X, my_xticks)
    elif (X[-1] - X[0] < 30):
        ticks = range(X[0],X[-1]+1)
        plt.xticks(ticks, ticks) 
    plt.xlabel("Outcomes")
    plt.ylabel("Probability")
    plt.title(title)
    plt.show()
    
# Example of use

draw_distribution_from_outcomes([1,4,3,5,4,6,2,4,3,5,4])
```

### Problem 7

How many integers in the range 1 .. 10000 (inclusive) are NOT divisible by 6, 7, or 8? (Hint: use Inclusion/Exclusion. The number of integers divisible by n would be  10000/n  .)

**Solution:**

Let $N_k$ be number of integers in [1..10,000] divisible by n, i.e., 

$$N_k = \left\lfloor\frac{10000}{k}\right\rfloor . $$

and let $N_{j,k} = N_j\cap N_k.$  

 The basic idea here is to remove the numbers divisible by each of 6,7,8, and then add back those divisible by two of them, then remove those divisible by all 3. 
 
 However, we can't use $6\cdot 8 = 48$ to remove the numbers divisible by 6 and 8, because for example 24 is divisible by both 6 and 8, but obviously 24 is not divisible by 48.  We have to use the least common multiple! 
 
Let `lcm(x,y)` be the least common multiple of x and y. Then

$$\begin{aligned}
       N_6 &= \left\lfloor\frac{10000}{6}\right\rfloor & \quad &= 1666   \\       
       N_7 &= \left\lfloor\frac{10000}{7}\right\rfloor & \quad &= 1428   \\ 
       N_8 &= \left\lfloor\frac{10000}{8}\right\rfloor & \quad &= 1250   \\ 
       N_{6,7} &= \left\lfloor\frac{10000}{lcm(6,7)}\right\rfloor &= \left\lfloor\frac{10000}{42}\right\rfloor  &= 238  \\ 
       N_{6,8} &= \left\lfloor\frac{10000}{lcm(6,8)}\right\rfloor &= \left\lfloor\frac{10000}{24}\right\rfloor &= 416  \\ 
       N_{7,8} &= \left\lfloor\frac{10000}{lcm(7,8)}\right\rfloor &= \left\lfloor\frac{10000}{56}\right\rfloor &= 178  \\ 
       N_{6,7,8} &= \left\lfloor\frac{10000}{lcm(6,7,8)}\right\rfloor &= \left\lfloor\frac{10000}{168}\right\rfloor &= 59  \\ 
\end{aligned}$$


Therefore 

$$N = N  N(A) - N(B) - N(C)] + N(AB) + N(AC) + N(BC)  N(ABC),$$

so 

$$N = 10000  1666 - 1428 - 1250 + 238 + 416 + 178  59 = 6429.$$

Therefore there are $\boxed{6429}$ integers between 1 and 10000 that are not divisible by 6, 7, or 8.
