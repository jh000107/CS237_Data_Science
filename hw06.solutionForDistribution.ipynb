## Problem Six

Consider an Ethernet with N nodes. During lull periods of time, no nodes are broadcasting a message; during such lull periods, a particular node will send a message with probabiliy $p$, independently of the other nodes. However, if 2 or more nodes send a message simultaneously, we have a collision and the messages are corrupted and transmission stops. A node whose message has been cancelled in this way will try again until the message is successfully transmitted. 

How many **times** on average will a given node attempt to transmit a message until it is successfully transmitted? 

But there is a subtlety here: Does "times" in the question mean 

<blockquote>
(i) "how many periods" (in which case you have to take account of the probability that the particular node attempted to send a message) or 

(ii) "attempts" (in which case the particular node has already tried, and the probability is 1.0 that the attempt was made)?
</blockquote>

(A)  Use assumption (i) to solve the problem. 

(B) Use assumption (ii) to solve the problem. 

<strong> Solution: </strong>

(A) In this case, we are watching from the outside, and counting periods until
a given node successfully transmits it message. 

The probability in any particular period that the station will attempt and successfully transmit or retransmit a message is $p(1p)^{N1}$.
This is because for the station to successfully transmit or retransmit its message, none of the
other stations should transmit messages in the same period. The number of periods until the given node successfully transmits its message is geometric with parameter $p(1p)^{N1}$.
Therefore, on average, the periods until success is 

$$\frac{1}{p(1p)^{N1}}.$$

(B) In this case, we watch this from the point of view of the node, and count
the number of attempts, not the number of periods. Any time
it attempts, it's probability of success is $(1-p)^{N-1}$, since all the
other nodes have to not transmit during this period. Therefore, if we
count attempts, this is $Geometric((1-p)^{N-1})$ and the expected
number of attempts until success is 

$$\frac{1}{(1p)^{N1}}.$$

<strong> Solution: </strong>  

Let $X$ be the number of rolls until Wayne gets a 2. Clearly, for $1\le k\le 6$, $P(X=k)={1\over 6}$  

Let $Y$ be the number of rolls of the die
until Trevor rolls an odd number; clearly $Y\sim Geometric(1/2)$.  

We seek $P(Y\gt X)$. The probability that Trevor has to roll more than $k$ times is:

$$P(Y\gt k) = (0.5)^k$$

since for $Y\gt k$, he must have an even number on each of the first $k$ rolls. 

If Wayne has to roll the die $k$ times and Trevor has to roll more than $k$, then the probability is  $P(X=k)\cdot P(Y\gt k)$, and we must sum over all possible $k$, so


$$\begin{aligned}
      \sum_{k=1}^\infty\, P(X=k)\cdot P(Y\gt k)  &= \sum_{k=1}^\infty\,\Bigl(\frac{1}{2}\Bigr)^k \cdot  
               \Bigl(\frac{5}{6}\Bigr)^{k-1} \cdot \frac{1}{6}\\[3pt]
           &= \frac{6}{5}\cdot \frac{1}{6}\cdot \sum_{k=1}^\infty\,\Bigl(\frac{5}{12}\Bigr)^{k}\\[3pt]
           &=    \frac{1}{5} \cdot  \frac{\frac{5}{12}}{1 - \frac{5}{12}}\\[3pt]
           &= \frac{1}{7}\\
\end{aligned}$$

## Problem Nine

Suppose a kennel has N cages for N dogs, and each dog wears a collar with his/her name, and each cage also has a label with the name of the dog on it. 
The owner of the kennel takes all the dogs out for exercise, and then
randomly lets them enter the cages, one dog per cage (so he pays no attention to the names).  

Let $X$ = the number of dogs who end up in their own cages. 

Find $E(X)$ and show all work. 

Hint: Define N Bernoulli "indicator" random variables, $I_1,\ldots I_N$, where success for $I_k$ is  dog $k$ ending up in
its own cage, and use the linearity of expectation. 

**Solution:**

Let $X$ = the number of dogs that end up in their own cages. 
We can most easily compute $E(X)$ by noting that


$$X = I_1 + I_2 + \cdots + I_N$$

where

\begin{equation}
        I_k = \begin{cases}
                   1               & \text{if dog $k$ ends up in the correct cage}\\
                   0 & \text{otherwise}
               \end{cases}
    \end{equation}

Now, since the $k^{th}$ dog is equally likely to end up in any of the $N$ cages,
each of the indicator RVs $I_k$ is distributed as $Bernoulli(1/N)$
and $E(I_k) = 1/N$. 

By the linearity of expectation, we have

$$E(X)\,=\, E(I_1) + E(I_2) + \cdots + E(I_N)$$

and so $E(X) = 1$.
